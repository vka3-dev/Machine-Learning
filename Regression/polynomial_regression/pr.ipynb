```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = 4 * np.random.rand(100,1) - 2
y = 4 + 2*x +5 * x**2 + 5 * np.random.rand(100, 1)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
df = pd.DataFrame({
    'x': x.flatten(),
    'y': y.flatten()
})
```

```python
df.to_csv("example_data.csv", index=False)
```

```python
reg = LinearRegression()
reg.fit(df[['x']], y)
```

```python
reg.predict([[2]])
```

```python
df['y_pred'] = reg.predict(df[['x']])
```

```python
plt.scatter(df['x'], df['y'], label='data_point')
plt.plot(df['x'], df['y_pred'], color='red', label='Regression Line')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly_features.fit_transform(df[['x']])
```

```python
reg.fit(x_poly, y)
```

```python
y_poly_pred = reg.predict(x_poly)
```

```python
sorted_idx = np.argsort(df['x'].values.flatten())
x_sorted = df['x'].values.flatten()[sorted_idx]
y_sorted = y[sorted_idx]
y_poly_sorted = y_poly_pred[sorted_idx]
```

```python
plt.scatter(df['x'], y, label='Data')

plt.plot(x_sorted, y_poly_sorted,color='red', linewidth=2,  label='Polynomial Fit')

plt.legend()
plt.show()
```
