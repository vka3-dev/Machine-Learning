```python
import pandas as pd
df = pd.read_csv("weight-height.csv")
df.head()
```

```python
q1 = df.Height.quantile(0.25)
q3 = df.Height.quantile(0.75)
wq1 = df.Weight.quantile(0.25)
wq3 = df.Weight.quantile(0.75)
IQR = q3 - q1
WIQR = wq3 - wq1
upper_limit = q3 + 1.5*IQR
lower_limit = q1 - 1.5*IQR
w_upper_limit = wq3 + 1.5*IQR
w_lower_limit = wq1 - 1.5*IQR
```

```python
df_with_no_outliers = df[(upper_limit > df.Height) & (df.Height > lower_limit) & (df.Weight > w_lower_limit) & (df.Weight < w_upper_limit)]
df_with_no_outliers
```
